```python
import re
import json
import requests
from bs4 import BeautifulSoup
```

```python
WORDLE_URL = 'https://www.nytimes.com/games/wordle/index.html'
OUTPUT = 'data/word_list.csv'
```

```python
res = requests.get(WORDLE_URL)
wordle_page = BeautifulSoup(res.content)
wordle_page
```

```python
js_files = [i.get('src') for i in wordle_page.find_all('script', type='text/javascript') if i.get('src')]
js_files
```

```python
pattern = r'\[[^\]]*?\bsnare\b[^\]]*?\]'
pattern = re.compile(pattern)

for js_file_url in js_files:
    res = requests.get(js_file_url)
    js_file = str(res.content)
    
    matches = re.findall(pattern, js_file)
    if matches:
        word_list = json.loads(matches[0])
        break

len(word_list)
```

```python
with open(OUTPUT, 'w') as f:
    f.write('\n'.join(word_list))
```
